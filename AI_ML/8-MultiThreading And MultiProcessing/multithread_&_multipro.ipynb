## MultiThreading
Q:-when to use multi threading\

ans:-

i). I/O-bound tasks:Tasks that spend more time waiting for I/O operations (e.g, file operations, network requests).

ii). Concurrent execution: when you want to improve the throughput of your application by performing multiple operations concurrently.

```python
import threading
import time
```

## single procees at a time executing and another one is waiting 

```python
def print_numbers():
    for i in range(5):
        time.sleep(2)   ## assume there any io operations running(here we just waiting this operation to done to move to next one)
        print(f"Numbers :{i}")

def print_my_letters(): ## waiting for above code to execute
    for letter in "abcde":
        time.sleep(2)
        print(f"Letter : {letter}")

t=time.time()

print_numbers()
print_my_letters()

finished_time=time.time()-t
print(finished_time)
```

## MultiThreading

time reduce from 20 sec to 10 sec

```python
def print_numbers():
    for i in range(5):
        time.sleep(2)   
        print(f"Numbers :{i}")

def print_letters():
    for letter in "abcde":
        time.sleep(2)
        print(f"Letter : {letter}")

## create 2 Threads
t1=threading.Thread(target=print_numbers)
t2=threading.Thread(target=print_letters)

t=time.time()

## start the thread
t1.start()
t2.start()

## wait for the threads to complete (after then print the total time)
t1.join()
t2.join()
finished_time=time.time()-t
print(finished_time)
```

## Multiprocessing



allows to create processes that run in parallel

Q:-when to use it....?

ans:-

i).CPU-Bound Tasks:- Tasks that are heavy on CPU usage (e.g., mathematical computations,data processing)

ii).Parallel Execution:- Multiple cores of the CPU



```python
import multiprocessing
import time 

def square_numbers():
    for i in range(5):
        time.sleep(1)
        print(f"Square : {i*i}")

def cube_numbers():
    for i in range(5):
        time.sleep(1.5)
        print(f"Cube : {i*i*i}")
if __name__=="__main__":
    ##create 2 process
    p1=multiprocessing.Process(target=square_numbers)
    p2=multiprocessing.Process(target=cube_numbers)

    t=time.time()
    ## strat the process
    p1.start()
    p2.start()

    ##wait for the process to complete
    p1.join()
    p2.join()

    finished_time=time.time()-t
    print(finished_time)
```
