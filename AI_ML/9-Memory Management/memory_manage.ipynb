## Refrence counting

Refernce counting is the primary method python uses to manage memory. Each object in python maintains a count of refernces pointing to it. when the refernce count droprs to zero, the memory occupied by the object is deallocated.

## Garbage collection

Python includes a cyclic garbage collector to handle refernce cycles. Reference cycles occur when objects reference each other , preventing their reference counts from reaching zero.

````python
import gc
## wanna enable garbage collection
gc.enable()
````

````python
## for disabling
gc.disable()
````

````python
gc.collect()
````

````python
### get garbage collection stats
print(gc.get_stats())
````

````python
## get unreachable objects
print(gc.garbage)
````

## Memory management best practices
1. Use Local Variables: Local variables have a shorter lifespan and are freed sooner than global variables.
2. Avoid Circular References : Circular references can lead to memory leaks if not properly managed.
3. Use generators : Generators produce items one at a time and only keep one itme in memory at a time,making them memory efficient.
4. Explicitly Delete Objects: use the del statement to delete  variables and objects explicitly 
5. Profile Memory Usage: use memory profiling tools like tracemalloc and memory profiler to identify memory leaks and optimize memory usage.

````python
import gc 

class MyObject:
    def __init__(self,name):
        self.name=name
        print(f"Object {self.name} created")
    def __del__(self):
        print(f"Object {self.name} deleted")

## create circular reference 
obj1=MyObject("obj1")
obj2=MyObject("obj2")
obj1.ref=obj2
obj2.ref=obj1

del obj1
del obj2

## Manually trigger the garbage collection
gc.collect()

 
````

del obj1  # removes the variable name obj1
del obj2  # removes the variable name obj2
```

Here's the important part — **`del` doesn't immediately delete the object from memory.**

It just removes the variable name. But the objects still exist in memory because they're still referencing each other!
```
obj1 variable gone ❌
obj2 variable gone ❌
But objects still in memory pointing to each other 👇
[object1] ←→ [object2]  # still alive in memory!
```

So `__del__` does NOT get called immediately here.

---

**Step 4 — Garbage Collector cleans up:**

Python's garbage collector detects this circular reference and cleans it up. Only THEN `__del__` gets called. (__del__ is a special function)

---

**Final output:**
```
Object obj1 created
Object obj2 created
Object obj1 deleted
Object obj2 deleted

## Generators for memory efficeiency

Generators allow you to produce items one at a time, using memory efficiently by only keeping one item in memory at a time.

````python
def generate_numbers(n):
    for i in range(n):
        yield i

## using the generator
for num in generate_numbers(100000):
    print(num)
    if(num>10):
        break
````

````python
## Profiling Memory usage with tracemalloc

import tracemalloc

def create_list():
    return [i for i in range(10000)]

def main():
    tracemalloc.start()

    create_list()

    snapshot = tracemalloc. take_snapshot()
    top_stats = snapshot.statistics('lineno')

    print("[ Top 10 ]")
    for stat in top_stats[:10]:
        print(stat)
````

````python
main()
````
