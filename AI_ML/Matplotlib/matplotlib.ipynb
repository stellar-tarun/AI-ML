## Data Visualization With Matplotlib

Matplotlib is a powerfull plotting library for Python that enables the creation of static,animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

build multiple plots

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5)) ##is fig size ke ander subplot me 
## for first graph
plt.subplot(1,2,1) ## 1 row me 2 coloumn and first place par yeh graph banega
plt.plot(x,y1,color='green')
plt.title("Plot 1")
```

subplot ka visualization

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[0.1,0.009,0.7,0.8,1]

plt.figure(figsize=(9,5))
## for first graph
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")
# plt.figure(figsize=(9,5)) agar yeh add ki toh fir ek aur graph ban jayega fir uske according niche wale subplot banega different (without this there will be two graoh in single row)

## for second graph
plt.subplot(1,2,2)
plt.plot(x,y1,color='green')
plt.title("Plot 2")
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[0.1,0.009,0.7,0.8,1]

plt.figure(figsize=(9,5))
## for first graph
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")
## for second graph
plt.subplot(2,2,2)
plt.plot(y1,y2,color='green')
plt.title("Plot 2")
## for third graph
plt.subplot(2,2,3)
plt.plot(y2,y1,color='green')
plt.title("Plot 3")
## for fourth graph
plt.subplot(2,2,4)
plt.plot(y1,x,color='green')
plt.title("Plot 4")
```

## Histograms

Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.

```python
data=[1,2,3,4,4,5,6,7,7,8,9,10]

plt.hist(data,bins=5,color='orange',edgecolor='black')
```

Create a scatter plot

```python
x=[1,2,3,4,5]
y=[2,3,4,5,6]
plt.scatter(x,y,color='black',marker='x')
```

pie chart

```python
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) ## move out the first slice
plt.pie(sizes,labels=labels)
```

```python
plt.pie(sizes,labels=labels,colors=colors)
```

for percentage


```python
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%")
```

Shadow (will create a lil bit shadow)

```python
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

explode (move out the first slice)

```python
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.3,0.3,0,0) ## move out the first slice (large the no.->larger the slice will be)
plt.pie(sizes,labels=labels)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

Sales Data visualization

```python
import pandas as pd
sales_data_df=pd.read_csv('sales_data.csv', encoding='latin1')
sales_data_df.head(5)

##Some CSV files are saved with a different encoding (like latin1 or cp1252) instead of UTF-8. When Pandas tries to read it with UTF-8 it fails on those special characters.
##latin1 encoding accepts almost all characters so it fixes the error.
```

```python
 sales_data_df.info()
```

 plot total sales by products

```python
total_sales_by_product=sales_data_df.groupby('ORDERNUMBER')['SALES'].sum()
total_sales_by_product
```

```python
total_sales_by_product.plot(kind='bar', color='pink',title="product category")
```

plot sales trend over time

```python
sales_trend=sales_data_df.groupby('ORDERDATE')['SALES'].sum().reset_index()
plt.plot(sales_trend['ORDERDATE'],sales_trend['SALES']) ##x axis-> orderdate
                                                        ##y axis-> sales
```
