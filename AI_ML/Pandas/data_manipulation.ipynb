filling missing values with the mean of the coloumn

```python
df['Ranking']=df['total_power'].fillna(df['total_power'].mean())
df
```

Renaming coloumns

```python
df.dtypes
```

```python
df=df.rename(columns={'name': 'heroes_name' })
df.head()
```

Changes data types

```python
df['new_total_power']=df['total_power'].astype(float) #it will gonna create at the last
df.head()
```

```python
df['new_total_power']=df['total_power'].fillna(df['Value'].mean()).astype(float) #it will gonna create at the last
df.head()
```

```python
df['New_total_power']=df['total_power'].apply(lambda x:x*2)
df.head()
```

Grouping Coloumn 

mean

```python
grouped_mean=df.groupby('heroes_name')['total_power'].mean()
print(grouped_mean)
```

sum

```python
grouped_sum=df.groupby(['heroes_name','species'])['total_power'].sum()
print(grouped_sum)
```

Aggregate multiple Functions

```python
grouped_age=df.groupby('species')['total_power'].agg(['mean','sum','count'])
grouped_age
```

Merging and joining DataFrames

create sample DataFrames

```python
df1=pd.DataFrame({'key' : ['A','B','C'] , 'Value1':[1,2,3]})
df2=pd.DataFrame({'key' : ['A','B','D'] , 'Value1':[4,5,6]})
```

```python
df1
```

```python
df2
```

merging this two dataframe

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

```python
pd.merge(df1,df2,on='key',how='inner')
```

inner -> only matching rows 

left -> all rows from left df

right -> all rows from right df

outer ->all rows from both

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,on='key',how='right')
```

```python
pd.merge(df1,df2,on='key',how='outer')
```
