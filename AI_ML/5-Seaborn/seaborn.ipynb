## Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlibn that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

Basic ploting with seaborn

```python
tips=sns.load_dataset('tips')
tips ##in built data set of resturant in seaborn library
```

create a scatter plot

```python
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of total bill")
```

line plot

```python
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("line plot of total bill")
plt.show()
```

Categorical plots (category variable->sex,smoker,day,time)

Bar plot

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

Box plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

Boxen plot

```python
sns.boxenplot(x='day',y='total_bill',data=tips)
```

Violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

Histograms

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False)
```

Kde plot

```python
sns.kdeplot(tips['total_bill'],shade=10) ##error jan bujh kar
```

```python
sns.kdeplot(tips['total_bill'],fill=True)
```

PairPlot

```python
sns.pairplot(tips)
```

Heatmap

```python
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
