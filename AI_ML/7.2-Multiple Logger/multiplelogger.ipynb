## Logging with Multiple Loggers

you can create multiple loggers for different parts of your application


```python
import logging
```

```python
## create a logger for module 1

logger1=logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

## create a logger for module 2

logger2=logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

## configure logging settings
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
## log message with different loggers

logger1.debug("This is debug message for module1")
logger2.warning("this is a warning message for module2")
logger2.error("this is an error message")
```

In the above code u r seeing that after the time there module 1 and module2 is written its showcasing that from which module this content get print but in previous code u will see there were root was written
