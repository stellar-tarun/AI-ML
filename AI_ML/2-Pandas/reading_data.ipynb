## Reading Data from different sources

json format

```python
import pandas as pd 
from io import StringIO
Data= '{"employee_name": "James","email":"james@gmail.com","job_profile":[{"title1":"Team Lead","title2":"Sr. Developer"}]}'
pd.read_json(StringIO(Data))
```

```python
df=pd.read_json(StringIO(Data))
df
```

```python
df.to_json()
```

```python
df.to_json(orient='index')
```

```python
df.to_json(orient='records')
```

pickle file

A pickle file is a file created using Python's pickle module to store a serialized Python object structure as a byte stream. This process, called pickling, allows complex Python objects (like lists, dictionaries, or even custom class instances) to be saved to disk and fully reconstructed later in another Python script, a process called unpickling

```python
df_excel=pd.read_excel('Book1.xlsx')
```

```python
df_excel.to_pickle('df_excel')
```

```python
##for reading this pickle file

pd.read_pickle('df_excel')
```
