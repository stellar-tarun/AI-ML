## Panda-DataFrame And Series

Pandas is a powerful data manipulation libraray in Python, widely used for data analysis and data cleaning.It provides two primary data structures. Series and DataFrame. A series is a one - Dimensional array-like object,while a DataFrame is a two - Dimensional , size-mutable,and potentially heterogenous tabular data structure with labeled axes (rows and columns).

DataFrame

create a dataframe from a dictionary of list

```python
data={
    'Name':['tarun','utkarsh','amit'],
    'Age':[21,20,22],
    'City':['Banglore','New York','Florida']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

Remove a coloumn

```python
df.drop('Occupation') ##{KeyError: "['Occupation'] not found in axis"} -> you
##will find this error over here,its bcoz   its going by default in row(axis=0) not in coloumn so it will find 0 ,1 ,2 but when we assign axis=1(for coloumn) then it will find name,age,height,occupation 
```

```python
df.drop('Occupation',axis=1)
```

inplace (what it do, is just that it remove that coloumn permanently)

```python
df ## (here you will find that table is not delete)
```

```python
df.drop('Occupation',axis=1,inplace=True) ##not it will remove permanently
```
