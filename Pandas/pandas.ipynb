## Panda-DataFrame And Series

Pandas is a powerful data manipulation libraray in Python, widely used for data analysis and data cleaning.It provides two primary data structures. Series and DataFrame. A series is a one - Dimensional array-like object,while a DataFrame is a two - Dimensional , size-mutable,and potentially heterogenous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
```

pandas is 1D array like object that can hold any data type. It is similar to a coloumn in a table.

```python
data=[1,2,3,4,5]
series=pd.Series(data)  ## bada wala S
print("Series \n",series)
```

create a series from dicitonary

```python
data={'a':1,'b':2,'c':3,'d':4}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[1,2,3,4,5]
index=['a','b','c','d','e']
pd.Series(data,index=index)
```

```python
pd.Series(index,index=data)
```

DataFrame

create a dataframe from a dictionary of list

```python
data={
    'Name':['tarun','utkarsh','amit'],
    'Age':[21,20,22],
    'City':['Banglore','New York','Florida']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

Create a data frame from a list of dictonaries

```python
data=[
    {'Name':'Tarun','Age':21,'Height':6.2},
    {'Name':'Utkarsh','Age':20,'Height':5.9},
    {'Name':'Amit','Age':22,'Height':5.7}
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
